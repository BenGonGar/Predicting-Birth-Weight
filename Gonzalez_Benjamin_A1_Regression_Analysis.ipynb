```python
# Checking new nulls
print(bw_test3.isnull().any(axis = 0))

#fixing new nulls
bw_test3['log_cigs'].fillna(0, inplace=True)
bw_test3['log_drink'].fillna(0, inplace=True)

# Checking new fix
print(bw_test3.isnull().any(axis = 0))
```

```python
#featuring
#bw_test3['if_over35_mage'] = 0 if bw_test3['mage'] < 35 else 1

bw_test3["if_over35_mage"] = (bw_test3["mage"] >= 36 ) * 1 #https://academic.oup.com/aje/article/186/11/1219/3858068

#did not improve
#bw_test3["if_over35_fage"] = (bw_test3["fage"] >= 36 ) * 1 #https://med.stanford.edu/news/all-news/2018/10/older-fathers-associated-with-increased-birth-risks.html

bw_test3["if_over2_npvis"] = (bw_test3["npvis"] >= 3 ) * 1 #https://medicine.yale.edu/news-article/prenatal-care-reduces-preterm-birth-and-low-birth-weight/

#did not improve
#bw_test3["if_over14_drink"] = (bw_test3["npvis"] >= 14 ) * 1 #https://pubmed.ncbi.nlm.nih.gov/6471316/

# looping to make x-variables suitable for statsmodels
for val in bw_test3:
    print(f"{val} +")

# Checking the results for every column
for columns in bw_test3:
    print(f"""{columns}""")
    print(bw_test3.loc[:,columns].value_counts(normalize = False,sort = True, ascending = False))  
```

```python
# Step 1: build a model
lm_best = smf.ols(formula =  
"""
log_bwght ~
cigs +
drink +
if_over35_mage +
if_over2_npvis 
""",              
data = bw_test3)


# Step 2: fit the model based on the data
results = lm_best.fit()


# Step 3: analyze the summary output
print(results.summary())
```

```python
# preparing explanatory variable data
bw_data   = bw_test3.drop([
"bwght",
"mage",
"log_bwght",
"meduc",
"monpre",
"male",
"npvis",
"fage",
"feduc",
"mwhte",
"mblck",
"moth",
"fwhte",
"fblck",
"foth",
"log_mage",
"log_npvis",
"log_meduc",
"log_monpre",
"log_fage",
"log_feduc",
"log_cigs",
"log_drink"
],axis = 1)


# preparing response variables
bw_target = bw_test3.loc[ : , "bwght"]
log_bw_target = bw_test3.loc[ : , "log_bwght"]


# preparing training and testing sets (all letters are lowercase)
x_train, x_test, y_train, y_test = train_test_split(
            bw_data,
            bw_target,
            test_size = 0.25,
            random_state = 219)

# preparing training and testing sets for log (all letters are lowercase)
x_train_log, x_test_log, y_train_log, y_test_log = train_test_split(
            bw_data,
            log_bw_target,
            test_size = 0.25,
            random_state = 219)


# checking the shapes of the datasets
print(f"""
Training Data
-------------
X-side: {x_train.shape}
y-side: {y_train.shape}


Testing Data
------------
X-side: {x_test.shape}
y-side: {y_test.shape}
""")

# checking the shapes of the log datasets
print(f"""
Training log Data
-------------
X-side: {x_train_log.shape}
y-side: {y_train_log.shape}


Testing log Data
------------
X-side: {x_test_log.shape}
y-side: {y_test_log.shape}
""")
```

```python
# INSTANTIATING a model object
ard_model = sklearn.linear_model.ARDRegression(normalize  = False)


# FITTING the training data
ard_fit = ard_model.fit(x_train_log, y_train_log)


# PREDICTING on new data
ard_pred = ard_fit.predict(x_test_log)


print('ARD Training Score:', ard_model.score(x_train_log, y_train_log).round(4))
print('ARD Testing Score :',  ard_model.score(x_test_log, y_test_log).round(4))


# saving scoring data for future use
ard_train_score = ard_model.score(x_train_log, y_train_log).round(4)
ard_test_score  = ard_model.score(x_test_log, y_test_log).round(4)


# displaying and saving the gap between training and testing
print('ARD Train-Test Gap :', abs(ard_train_score - ard_test_score).round(4))
ard_test_gap = abs(ard_train_score - ard_test_score).round(4)

# zipping each feature name to its coefficient
ard_model_values = zip(bw_data.columns, ard_fit.coef_.round(decimals = 5))


# setting up a placeholder list to store model features
ard_model_lst = [('intercept', ard_fit.intercept_.round(decimals = 2))]


# printing out each feature-coefficient pair one by one
for val in ard_model_values:
    ard_model_lst.append(val)
    

# checking the results
for pair in ard_model_lst:
    print(pair)
```

```python
# this is the exact code we were using before
x_train_STAND, x_test_STAND, y_train_STAND, y_test_STAND = train_test_split(
            x_scaled_df,
            log_bw_target,
            test_size = 0.25,
            random_state = 219)
```

```python
# creating lists for training set accuracy and test set accuracy
training_accuracy = []
test_accuracy = []


# building a visualization of 1 to 50 neighbors
neighbors_settings = range(1, 51)


for n_neighbors in neighbors_settings:
    # Building the model
    clf = KNeighborsRegressor(n_neighbors = n_neighbors)
    clf.fit(x_train_STAND, y_train_STAND)
    
    # Recording the training set accuracy
    training_accuracy.append(clf.score(x_train_STAND, y_train_STAND))
    
    # Recording the generalization accuracy
    test_accuracy.append(clf.score(x_test_STAND, y_test_STAND))


# plotting the visualization
fig, ax = plt.subplots(figsize=(12,8))
plt.plot(neighbors_settings, training_accuracy, label = "training accuracy")
plt.plot(neighbors_settings, test_accuracy,     label = "test accuracy")
plt.ylabel("Accuracy")
plt.xlabel("n_neighbors")
plt.legend()
plt.show()


# finding the optimal number of neighbors
opt_neighbors = test_accuracy.index(max(test_accuracy)) + 1
print(f"""The optimal number of neighbors is {opt_neighbors}""")
```

```python
# INSTANTIATING a model with the optimal number of neighbors
knn_stand = KNeighborsRegressor(algorithm = 'auto',
                                n_neighbors = opt_neighbors)



# FITTING the model based on the training data
knn_stand_fit = knn_stand.fit(x_train_STAND, y_train_STAND)



# PREDITCING on new data
knn_stand_pred = knn_stand_fit.predict(x_test_STAND)



# SCORING the results
print('KNN Training Score:', knn_stand.score(x_train_STAND, y_train_STAND).round(4))
print('KNN Testing Score :',  knn_stand.score(x_test_STAND, y_test_STAND).round(4))


# saving scoring data for future use
knn_stand_score_train = knn_stand.score(x_train_STAND, y_train_STAND).round(4)
knn_stand_score_test  = knn_stand.score(x_test_STAND, y_test_STAND).round(4)


# displaying and saving the gap between training and testing
print('KNN Train-Test Gap:', abs(knn_stand_score_train - knn_stand_score_test).round(4))
knn_stand_test_gap = abs(knn_stand_score_train - knn_stand_score_test).round(4)
```
